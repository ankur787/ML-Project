# Features Selection for model

```python
dfp_final.columns
```

```python
# Based on EDA these variable are relevent to the model
features = ['DAY', 'Tags_served', 'Impressions', 'Clicks', 'Revenue','Story', 'Position', 'ampTF','Actual_Revenue' ]
```

```python
dfp_model = dfp_final[features]
```

```python
dfp_model.shape
```

```python
dfp_model.columns
```

```python
dfp_model.info()
```

```python
# object Columns selection
dummy=dfp_model.select_dtypes('object')
```

```python
dummy.head()
```

```python
# Create dummies 
dummy= pd.get_dummies(dummy)
```

```python
dummy.head()
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
mms= MinMaxScaler()
```

```python
dfp_model_scaler =dfp_model.select_dtypes(['int64', 'float64'])
```

```python
dfp_model_scaler.loc[:, dfp_model_scaler.columns != "Actual_Revenue"] = mms.fit_transform(dfp_model_scaler.loc[:, dfp_model_scaler.columns !="Actual_Revenue"])
```

```python
dfp_model_scaler.head()
```

```python
dfp_model = dfp_model_scaler.join(dummy)
```

```python
dfp_model.head()
```

```python
dfp_model.columns
```

```python
# Rearrange the oder
dfp_model =dfp_model[['DAY', 'Tags_served', 'Impressions', 'Clicks', 'Revenue', 'ampTF','Story_', 'Story_eighth', 'Story_fifth', \
       'Story_first', 'Story_fourth', 'Story_home', 'Story_ninth', 'Story_second', 'Story_section', 'Story_seventh', \
             'Story_sixth','Story_tenth', 'Story_third', 'Position_afterrelated','Position_bottom', 'Position_leaderboard', \
             'Position_middle','Position_top', 'Actual_Revenue']]
```

```python
from sklearn.model_selection import train_test_split
```

```python
dfp_model.head()
```

```python
dfp_model.shape
```

```python
X= dfp_model.iloc[:,:24]
```

```python
y= dfp_model['Actual_Revenue']
```

```python
print(X.shape), print(y.shape)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25)
```

```python
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
from sklearn import linear_model
from sklearn.metrics import mean_squared_error, r2_score
```

```python
# Create linear regression object
regr = linear_model.LinearRegression()
```

```python
# Train the model using the training sets
regr.fit(X_train, y_train)
```

```python
# Make predictions using the testing set
dfp_y_pred = regr.predict(X_test)
```

```python
# The coefficients
print('Coefficients: \n', regr.coef_)
```

```python
# The mean squared error
print("Mean squared error: %.2f"
      % mean_squared_error(y_test, dfp_y_pred))
```

```python
# Explained variance score: 1 is perfect prediction
print('Variance score: %.2f' % r2_score(y_test, dfp_y_pred))
```

```python
plt.scatter(y_test, dfp_y_pred)
plt.xlabel('True Values')
plt.ylabel('Predictions')
```
